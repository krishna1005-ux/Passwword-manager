```python
from tkinter import*
from tkinter import messagebox
krish=Tk()
krish.geometry("400x400")

krish.config(bg="#f0f0f0")
#title
krish.title("üîêPassword Manager")
#label
title=Label(krish,text="üîê Password Manager",fg="#000400",font="arial 16 bold")
title.pack(pady=10)

#frame
frame=Frame(krish)
frame.pack()
show_password=BooleanVar()
theme_mode=BooleanVar()

def apply_theme():
    if theme_mode.get():
        bg_color = "#1e1e1e"
        fg_color = "#ffffff"
        entry_bg = "#2e2e2e"
    else:
        bg_color = "#f0f0f0"
        fg_color = "#000000"
        entry_bg = "#ffffff"

    # Apply window and frame background
    krish.config(bg=bg_color)
    form_frame.config(bg=bg_color)
    form_btn.config(bg=bg_color)
    title.config(bg=bg_color, fg=fg_color)

    # Update labels and entries
    for widget in form_frame.winfo_children():
        if isinstance(widget, Label):
            widget.config(bg=bg_color, fg=fg_color)
        elif isinstance(widget, Entry):
            widget.config(bg=entry_bg, fg=fg_color, insertbackground=fg_color)

    # Update buttons and checkbutton
    for widget in form_btn.winfo_children():
        widget.config(bg=bg_color, fg=fg_color)

    theme_toggle.config(bg=bg_color, fg=fg_color, selectcolor=bg_color)

#save function
def save():
    site=site_entry.get()
    username=username_entry.get()
    password=password_entry.get()
    print("your data is")
    print("Site:",site)
    print("username",username)
    print("password",password)

    site_entry.delete(0,END)
    username_entry.delete(0,END)
    password_entry.delete(0,END)

    with open("password.txt","a")as file:
        file.write(f"{site} | {username} | {password}\n")

    messagebox.showinfo("success","password saved successfully!")


    

#hide password function
def toggle_password():
    if show_password.get() is True:
        password_entry.config(show="")
    else:
        password_entry.config(show="*")

#save data window function
def view():
    win=Toplevel()
    win.title("Your Saved Passwords")
    win.geometry("400x400")
    
    text=Text(win,wrap=WORD)
    text.pack(expand=True,fill=BOTH)

    try:
        with open("password.txt","r") as file:
            content=file.read()
            text.insert(END,content)
    except FileNotFoundError:
        text.insert(END,"No password saved yet.")

#form fram
form_frame=Frame(krish,bg="#f0f0f0")
form_frame.pack(pady=10)
#labels
label1=Label(form_frame,text="üåê Site:",font="arial 16 bold").grid(row=0,column=0,sticky=W,padx=5,pady=5)
site_entry=Entry(form_frame, width=30)
site_entry.grid(row=0, column=1, padx=5)

label2=Label(form_frame,text="üë§ UserName:",font="arial 16 bold",fg="#000400").grid(row=1,column=0,sticky=W,pady=5)
username_entry=Entry(form_frame, width=30)
username_entry.grid(row=1,column=1,pady=5)

label3=Label(form_frame,text="üîë Password:",font="arial 16 bold",fg="#000400").grid(row=2,column=0,sticky=W,pady=5)
password_entry=Entry(form_frame,show="*",width=30)
password_entry.grid(row=2,column=1,pady=5)



#form button
form_btn=Frame(krish,bg="#f0f0f0")
form_btn.pack(pady=10)

#savebutton
save_button=Button(form_btn,text="üíæ SAVE",command=save,width=12,font="arial 8 bold")
save_button.grid(row=0,column=0,padx=5,pady=5)

#new window
view=Button(form_btn,text="üìÇ View Password",command=view,width=15,font="arial 8 bold")
view.grid(row=0,column=1,padx=5,pady=5)

#checkbutton
show=Checkbutton(form_btn,text=" Show Password",variable=show_password,command=toggle_password,width=16,font="arial 8 bold")
show.grid(row=0,column=2,padx=5,pady=5)
#Label(win, text="Your Saved Passwords:", font=("Arial", 10, "bold")).pack(pady=5)

theme_toggle=Checkbutton(krish,text="üåô Dark Mode",variable=theme_mode,command=apply_theme,font="arial 12 bold")
theme_toggle.pack(pady=5)

apply_theme() # apply default (light) theme on start

krish.mainloop()
```
